# Testing video concatenation

```python
import os
os.environ["IMAGEIO_FFMPEG_EXE"] = "/usr/local/bin/ffmpeg"  # Update this path as needed

from moviepy.editor import VideoFileClip, ImageClip, AudioFileClip, concatenate_videoclips
import scipy
from IPython.display import Audio

from transformers import AutoProcessor, BarkModel
```

```python
clip1 = VideoFileClip("videos/waving.MOV")
clip2 = VideoFileClip("videos/fellas.MOV")
final_clip = concatenate_videoclips([clip1,clip2])
final_clip.write_videofile("concat_v0.mp4")
```

```python
processor = AutoProcessor.from_pretrained("suno/bark")
model = BarkModel.from_pretrained("suno/bark")

voice_preset = "v2/en_speaker_6"

inputs = processor("Hi! We've created a podlight for you all about AI. Enjoy!", voice_preset=voice_preset)

audio_array = model.generate(**inputs)
audio_array = audio_array.cpu().numpy().squeeze()
```

```python
sample_rate = model.generation_config.sample_rate
Audio(audio_array, rate=sample_rate)
```

```python
sample_rate = model.generation_config.sample_rate
scipy.io.wavfile.write("sample_intro.wav", rate=sample_rate, data=audio_array)
```

```python
# Load the audio file
audio_clip = AudioFileClip("sample_intro.wav")

# Create an ImageClip object with the duration set to the audio clip's duration
image_clip = ImageClip("logo.png", duration=audio_clip.duration)

# Set the audio of the ImageClip
video_clip = image_clip.set_audio(audio_clip)

# Write the result to a file (output file will be a video)
video_clip.write_videofile("sample_output_video.mp4", fps=2)
```
